```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

График №1 - Самый калорийный напиток

```python
with open('КБЖУ.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
foods = data['foods']
activities = data['activities']
```

```python
drinks = [f for f in foods if f['category'] == 'напитки']
most_caloric = max(drinks, key=lambda x: x['calories'])

top_drinks = sorted(drinks, key=lambda x: x['calories'], reverse=True)[:5]

fig, ax = plt.subplots()

names = [d['name'] for d in top_drinks]
calories = [d['calories'] for d in top_drinks]

bars = ax.bar(names, calories)
ax.set_title('Топ-5 самых калорийных напитков', fontweight='bold')
ax.set_ylabel('Калории на 100мл')
ax.tick_params(axis='x', rotation=45)

for bar, val in zip(bars, calories):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2., height + 2,
        f'{val} ккал', ha='center')

plt.tight_layout()
plt.show()
```
