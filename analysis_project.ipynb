**Анализ данных**

```python
import json
import matplotlib.pyplot as plt
import numpy as np

```

```python
with open('КБЖУ.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

График №1 - Самый калорийный напиток

```python
all_drinks = []
for product in data['foods']:
    if product['category'] == 'напитки':
        all_drinks.append(product)

all_drinks.sort(key=lambda x: x['calories'], reverse=True)
top_5_drinks = all_drinks[:5]

names = []
calories = []
for drink in top_5_drinks:
    names.append(drink['name'])
    calories.append(drink['calories'])

plt.bar(names, calories)
plt.title('Самые калорийные напитки')
plt.ylabel('Калории в 100 мл')

for i in range(len(names)):
    plt.text(i, calories[i] + 2, str(calories[i]), ha='center')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Вывод: на графике можно увидеть, что самым калорийным напитком из представленных является вино, которое обгоняет ближайших конкурентов почти в 2 раза. На 100 мл приходится 83 ккал.

График №2 - Самые эффективные активности

```python
activities = data['activities']
activities.sort(key=lambda x: x['calories_per_hour'], reverse=True)

top5_activities = activities[:5]

names = []
calories = []
for activity in top5_activities:
    names.append(activity['name'])
    calories.append(activity['calories_per_hour'])

plt.bar(names, calories, color='green')
plt.title('Самые эффективные активности')
plt.ylabel('Калории в час')

for i in range(len(names)):
    plt.text(i, calories[i] + 10, str(calories[i]), ha='center')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Вывод: на графике можно увидеть, что больше всего сжигает калорий интенсивный бег, а также интенсивный велосипед и средний бег средний, значит при выборе активности для похудения есть смысл рассматривать их в первую очередь (однако не стоит забывать, что калории расчитаны на час и далеко не каждый неподготовленный человек сумеет выдержать достаточный темп для выполнения нормы и сжигания столько количества калорий, которое он хочет)

График №3 - Самое большое количество элементов

```python
foods = data['foods']
max_protein = max(foods, key=lambda x: x['protein'])
max_fat = max(foods, key=lambda x: x['fat'])
max_carbs = max(foods, key=lambda x: x['carbs'])

nutrients = ['Белки', 'Жиры', 'Углеводы']
values = [max_protein['protein'], max_fat['fat'], max_carbs['carbs']]
products = [max_protein['name'], max_fat['name'], max_carbs['name']]

plt.bar(nutrients, values, color=['pink'])
plt.title('Продукты с максимальным содержанием БЖУ')
plt.ylabel('Граммы элементов на 100г')

for i in range(len(nutrients)):
    plt.text(i, values[i] + 2, f'{products[i]}', ha='center')
plt.tight_layout()
plt.show()
```

Вывод: данный график был создан в развлекательных, информативных целях и мы можем увидеть, что в сахаре и масле подсолнечном максимальное количество углеводом и жира соответственно, что делает их абсолютными источниками веществ (однако мы знаем, что просто так их употреблять может быть опасно для организма, поэтому они лишь ингредиенты для блюд)

График №4 - Диаграмма БЖУ в Омлете

```python
foods = data['foods']
omlet = None
for food in foods:
    if food['name'] == 'омлет':
        omlet = food
        break

labels = ['Белки', 'Жиры', 'Углеводы']
sizes = [omlet['protein'], omlet['fat'], omlet['carbs']]

plt.pie(sizes, labels=labels)
plt.title(f'Состав омлета {omlet["calories"]} ккал на 100г')
plt.show()
```

Вывод: также в информациооных целях представленно разложение Омлета на БЖУ, где мы можем увидеть, что белки и жиры можно сказать в двоем делят состав блюда, что при поедании можно расчитывать на получение как и жиров, так и белков одинакого

График №5 - Минимальное количество калорий

```python
foods = data['foods']
foods.sort(key=lambda x: x['calories'])
top10_calorie = foods[:10]

names = []
calories = []
for food in top10_calorie:
    names.append(food['name'])
    calories.append(food['calories'])

plt.bar(names, calories, color='lightblue')
plt.title('Продукты с наименьшей калорийностью')
plt.ylabel('Калории на 100г')

for i in range(len(names)):
    plt.text(i, calories[i] + 2, str(calories[i]), ha='center')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Вывод: на графике можно увидеть самые низкокалорийные продукты, и очевидно, что это базовые напитки и овощи. Человек выбирающий еду для похудения может получить много полезного из этого списка.
